# InfraRiskAI EDA 01 - Real Data Overview

This notebook inventories the local real-data assets used by the dashboard and backend. It does not create synthetic fallback rows.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
DATA = ROOT / 'data'
RAW = DATA / 'raw'
PROCESSED = DATA / 'processed'

sources = json.loads((DATA / 'source_registry' / 'real_data_sources.json').read_text())['sources']
availability = json.loads((PROCESSED / 'real_data_availability_report.json').read_text())['sources']
pd.DataFrame(availability)[['source_id', 'name', 'status', 'records', 'message']]
```

```python
files = {
    'World Bank PPI': RAW / 'ppi' / 'ppi_projects.csv',
    'WDI macro': RAW / 'worldbank' / 'wdi_macro.csv',
    'CDS spreads': RAW / 'worldbank' / 'cds_spreads.csv',
    'National Bridge Inventory': RAW / 'nbi' / 'nbi_bridges.csv',
    'Yahoo Finance prices': RAW / 'market' / 'yahoo_finance_prices.csv',
}

summary = []
for name, path in files.items():
    if path.exists():
        df = pd.read_csv(path, low_memory=False)
        summary.append({'dataset': name, 'rows': len(df), 'columns': len(df.columns), 'path': str(path.relative_to(ROOT))})
    else:
        summary.append({'dataset': name, 'rows': 0, 'columns': 0, 'path': str(path.relative_to(ROOT))})
pd.DataFrame(summary)
```
