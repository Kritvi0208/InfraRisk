# InfraRiskAI EDA 03 - Risk Signals

Reviews real macro, market, and engineering signals available locally for credit and infrastructure risk modeling.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
RAW = ROOT / 'data' / 'raw'
wdi = pd.read_csv(RAW / 'worldbank' / 'wdi_macro.csv', low_memory=False)
cds = pd.read_csv(RAW / 'worldbank' / 'cds_spreads.csv', low_memory=False)
nbi = pd.read_csv(RAW / 'nbi' / 'nbi_bridges.csv', low_memory=False)
prices = pd.read_csv(RAW / 'market' / 'yahoo_finance_prices.csv', low_memory=False)
wdi.shape, cds.shape, nbi.shape, prices.shape
```

```python
display(wdi.head())
display(cds.head())
display(nbi.head())
display(prices.head())
```

```python
numeric_wdi = wdi.select_dtypes('number')
numeric_cds = cds.select_dtypes('number')
numeric_nbi = nbi.select_dtypes('number')

display(numeric_wdi.describe().T.head(20))
display(numeric_cds.describe().T.head(20))
display(numeric_nbi.describe().T.head(20))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
if not numeric_cds.empty:
    numeric_cds.iloc[:, : min(4, numeric_cds.shape[1])].plot(ax=axes[0], title='CDS / Spread Signals')
if not numeric_nbi.empty:
    numeric_nbi.iloc[:, : min(4, numeric_nbi.shape[1])].hist(ax=axes[1])
    axes[1].set_title('NBI Numeric Signal Distributions')
plt.tight_layout()
```
